The dust flux is a useful quantity, but what we really want is how does our estimate of the total dust injected into the atmosphere differ with the different detection schemes.

In order to calculate the total dust injected, we need not just the dust flux from individual devils but also their total areas AND their lifetimes. Strictly, we can't estimate the total dust injected into the atmosphere without knowing their diameters (or, more accurately, their surface areas) and lifetimes. 

The observed encounter durations ($\Gamma_{\rm obs}$ = SCLK_end - SCLK_start) are related to their diameters $D$ as:
$$
D^2 = \left( \Gamma_{\rm obs} \upsilon \right)^2 - 4b^2,
$$
where $\upsilon$ is the vortex's velocity and $b$ is the (unknown) closest approach distance between the sensor and vortex center. See Jackson et al. (2018 - https://ui.adsabs.harvard.edu/abs/2018Icar..299..166J/abstract) for details. If we assume $b \approx {\rm const.}$, we can just use the observed encounter durations to calculate a scaled diameter for each. Then we can use the lifetime-diameter scaling from Lorenz (2013 - https://www.sciencedirect.com/science/article/pii/S0019103513002960) to estimate relative lifetimes.
