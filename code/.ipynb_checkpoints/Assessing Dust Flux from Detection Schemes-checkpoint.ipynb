```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# Just some default settings
from utils import *

def calc_Neakrase_dust_flux(Delta_P, norm_factor=1e-4):
    # From Neakrase+ (2010 - https://ui.adsabs.harvard.edu/abs/2010Icar..206..306N/abstract)
    #
    # norm_factor = 1e-4 #kg/m^2/s
    # Delta P in Pascal
    
    return norm_factor*Delta_P**3.4

def calc_lifetime(diameter, norm_factor=40.):
    # From Lorenz (2013 - https://www.sciencedirect.com/science/article/pii/S0019103513002960)
    #
    # norm_factor = 40. # seconds
    # diameter in meters

    return norm_factor*diameter**(2./3)
```

The dust flux is a useful quantity, but what we really want is how does our estimate of the total dust injected into the atmosphere differ with the different detection schemes.

In order to calculate the total dust injected, we need not just the dust flux from individual devils but also their total areas AND their lifetimes. Strictly, we can't estimate the total dust injected into the atmosphere without knowing their diameters (or, more accurately, their surface areas) and lifetimes. 

The observed encounter durations ($\Gamma_{\rm obs}$ = SCLK_end - SCLK_start) are related to their diameters $D$ as:
$$
D^2 = \left( \Gamma_{\rm obs} \upsilon \right)^2 - 4b^2,
$$
where $\upsilon$ is the vortex's velocity and $b$ is the (unknown) closest approach distance between the sensor and vortex center. See Jackson et al. (2018 - https://ui.adsabs.harvard.edu/abs/2018Icar..299..166J/abstract) for details. If we assume $b \approx 0$, which has to be nearly true for us to detect the devil in the first place, we can just use the observed encounter durations to calculate a scaled diameter for each. Then we can use the lifetime-diameter scaling from Lorenz (2013 - https://www.sciencedirect.com/science/article/pii/S0019103513002960) to estimate relative lifetimes, which gives $\tau \propto D^{2/3}$.

Thus, the total dust $T$ injected into the atmosphere by a dust devil with a pressure dip $\Delta P$, diameter $D$, and lifetime $\tau$ is
$$
T = Q(\Delta P) \ \frac{\pi}{4} D^2\  \tau(D) \propto \Delta P^{3.4}\ \Gamma_{\rm obs}^2\ \Gamma_{\rm obs}^{2/3} = \Delta P^{3.4}\ \Gamma_{\rm obs}^{8/3}.
$$

```python
# Convert Delta P values to dust fluxes
fig = plt.figure(figsize=(6*aspect_ratio, 6))
ax = fig.add_subplot(111)

# Only encounters that show up as ground-truth
ind = ~np.isnan(data['gt_vortex_ind'])
# use norm_factor = 1 to make proportional
ground_truth_dust_fluxes = calc_Neakrase_dust_flux(np.abs(data['PRESSURE_DIFF'][ind]), norm_factor=1.)
# To normalize out diameter units, I will assume v = 1 m/s. This advection speed is arbitrary but reasonable.
advection_speed = 1. # m/s
approximate_diameter = (data['SCLK_end'][ind].astype(float) - data['SCLK_start'][ind].astype(float))*advection_speed
ground_truth_total_dust = ground_truth_dust_fluxes*approximate_diameter**(8./3)
ax.loglog(np.abs(data['PRESSURE_DIFF'][ind]), ground_truth_total_dust, ls='', marker='o', color=BoiseState_blue, markersize=10,
          label=r"$\sum T_{\rm GT}/10^9 = %.3g$" % (np.sum(ground_truth_total_dust)/1e9))

print(approximate_diameter)

# Only ground-truth detections that Wronkiewicz recovered
ind = ~np.isnan(data['associated_gt_vortex_ind'])
# use norm_factor = 1 to make proportional
recovered_dust_fluxes = calc_Neakrase_dust_flux(np.abs(data['PRESSURE_DIFF'][ind]), norm_factor=1.)
# To normalize out diameter units, I will assume v = 1 m/s. This advection speed is arbitrary but reasonable.
advection_speed = 1. # m/s
approximate_diameter = (data['SCLK_end'][ind].astype(float) - data['SCLK_start'][ind].astype(float))*advection_speed
recovered_total_dust = recovered_dust_fluxes*approximate_diameter**(8./3)
ax.loglog(np.abs(data['PRESSURE_DIFF'][ind]), recovered_total_dust, ls='', marker='o', color=BoiseState_orange, 
          label=r"$\sum T_{\rm SR}/10^9 = %.3g$" % (np.sum(recovered_total_dust)/1e9))

ax.tick_params(labelsize=24)
ax.set_xlabel(r'$\Delta P\,\left( {\rm Pa} \right)$', fontsize=36)
ax.set_ylabel(r'$T\,\left( {\rm Arb. Units} \right)$', fontsize=36)
ax.grid(True)
ax.legend(loc='best', fontsize=24)

plt.show()
fig.savefig("../figures/Total_Dust_Injected_norm_prob_0.95.jpg", dpi=300, bbox_inches="tight")
```
